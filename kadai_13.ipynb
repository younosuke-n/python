tips.csvを読み込み、以下を求めよ（tips.csvは講義資料のページのサンプルファイルに含まれている）。

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('./examples/tips.csv')
data
```

## 1

smokerを行とし、total_billの最高値と最低値を曜日ごとに列に示したデータフレームを作れ。

```python
data.pivot_table(['total_bill'], index='smoker', columns='day', aggfunc=['max', 'min'])
```

## 2

曜日、smokerを行の階層型インデックスとし、total_billの平均、標準偏差、最大値と最小値の差、の３列を表示したデータフレームを作れ。

```python
def peak_to_peak(arr):
    return arr.max() - arr.min()
```

```python
data.pivot_table(['total_bill'], index=['smoker', 'day'], aggfunc=['mean', 'std', peak_to_peak])
```

## 3

読み込んだデータフレームに、総支払額(total_bill)に占めるチップの額(tip)の割合を示す列tip_pctを追加せよ。

```python
data['tip_dict'] = data['tip'] / data['total_bill']
data
```

## 4

曜日ごとのtotal_billの合計とsizeの合計、さらにtip_pctの平均を示したデータフレームを作れ。

```python
data.pivot_table(['total_bill', 'size', 'tip_dict'], index='day', aggfunc={'total_bill': 'sum', 'size': 'sum', 'tip_dict': 'mean'})
```
