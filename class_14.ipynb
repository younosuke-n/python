# matplotlibによる可視化

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
%matplotlib notebook
```

## 練習問題1

```python
fig, axes = plt.subplots(2, 2)
data = np.arange(10)
```

```python
axes[0, 0].plot(np.sin(data))
axes[0, 1].plot(np.cos(data))
axes[1, 0].plot(np.power(data, 2))
```

## 練習問題2

```python
fig, axes = plt.subplots(2, 2)
data = np.arange(10)

axes[0, 0].plot(np.sin(data), marker='*')
axes[0, 1].plot(np.cos(data), marker='+')
axes[1, 0].plot(np.power(data, 2), marker='v')
```

## 練習問題3

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

x = np.linspace(-10, 10, 1000)
ax.plot(x, x+np.sin(x), color='black', label='x + sin(x)')
ax.plot(x, x*np.cos(x), linestyle='dashed', color='blue', label='x * cos(x)')

ticks = ax.set_xticks([-3*np.pi, -2*np.pi, -np.pi, 0, np.pi, 2*np.pi, 3*np.pi])
labels = ax.set_xticklabels(['-3pi', '-2pi', 'pi', 'O', 'pi', '2pi', '3pi'])
ax.legend(loc='best')
```

## 練習問題7

```python
def toInt(x):
    x = x.replace(',', '')
    return int(x)

def toFloat(x):
    x = x.replace(',', '')
    return float(x)
```
