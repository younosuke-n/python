## 練習問題1

何かしらのファイルを開き、何かしらのメッセージを書き込むが、すでにそのファイルが存在する場合に'filename already exists!'というメッセージを出力するコードを例外ハンドラを利用して書け。 filenameは開こうとしたファイルの名前である。

```python
poem = '''There was a young lady named Bright,
whose speed was far faster than light;
She started one day
In a relative way,
And returned on the previous night.'''
```

```python
fout = open('relativity.txt', 'wt')
fout.write(poem)
fout.close()
```

```python
def msg_write_file(msg, filename):
    try:
        fout = open(filename, 'xt')
        fout.write(msg)
        fout.close()
    except FileExistsError:
        print(filename, 'already exists!')
```

```python
filename1 = 'relativity.txt'
filename2 = 'relative.txt'
```

```python
msg_write_file(poem, filename1)
```

```python
msg_write_file(poem, filename2)
```

## 練習問題4

以下のようなcsvファイルをcsv.writerを用いて作成せよ。<br>
ヒント. writerow(('1', '2', '3'))のようにすると、１行ずつ書き出せる。

```python
import csv
```

```python
with open('onetwothree.csv', 'w') as fout:
    cout = csv.writer(fout, delimiter='|', lineterminator='\n')
    cout.writerow(('one', 'two', 'three'))
    cout.writerow(('1', '2', '3'))
    cout.writerow(('4', '5', '6'))
    cout.writerow(('7', '8', '9'))
    
!type onetwothree.csv
```

## 練習問題5

```python
import xml.etree.ElementTree as et
```

```python
tree = et.ElementTree(file='./examples/exercise5.xml')
root = tree.getroot()
```

```python
for child in root:
    for gchild in child:
        if gchild.tag == 'string' and gchild.attrib['key'] == 'concept:name':
                print('case ID:',gchild.attrib['value'])
        for ggchild in gchild:
            if ggchild.attrib['key'] == 'concept:name':
                print('\t',ggchild.attrib['value'])
            if ggchild.attrib['key'] == 'lifecycle:transition':
                print('\t',ggchild.attrib['value'])
```
