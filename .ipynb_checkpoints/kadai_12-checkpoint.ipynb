```python
import numpy as np
import pandas as pd
```

```python
frame = pd.DataFrame({'prefecture': ['Nagasaki', 'Nagasaki', 'Nagasaki', 'Fukuoka', 'Fukuoka', 'Fukuoka'], \
                     'city': ['Nagasaki', 'Sasebo', 'Isahaya', 'Fukuoka', 'Kitakyushu', 'Kurume'], \
                     'population': [429508, 255439, 138078, 1555508, 946338, 304705], \
                     'gdp': [15574, 7719, 6362, 77525, 37616, 14963]})
frame
```

```python
pd.options.display.float_format = '{:%}'.format

f = lambda x: x / frame.groupby('prefecture').sum().loc[x.loc['prefecture']]
frame.apply(f, axis='columns').filter(items=frame.groupby('prefecture').sum().columns)
```
